## Training Pipeline Functions

### Data Cleaning

```python
# Step 1 - Apply to test
def clean_missing(df_train, df_test, drop_pct=0.1):
    
    # Dealing with missing values
    # 1. Drop Columns if missing values is over drop_pct
    sum_of_miss = df_train.isnull().sum() / len(df_train)
    cols_with_miss_over_pct = sum_of_miss[sum_of_miss > drop_pct].index.tolist()
    df_train = df_train.drop(cols_with_miss_over_pct, axis=1)
    df_test = df_test.drop(cols_with_miss_over_pct, axis=1)
    
    # 2. Dealing with Garage variables
    df_train = df_train.drop(['GarageYrBlt'], axis=1)
    df_test = df_test.drop(['GarageYrBlt'], axis=1)
    fill_dict = {
        'GarageType': 'Ng',
        'GarageFinish': 'Ng',
        'GarageQual': 'Ng',
        'GarageCond': 'Ng'
    }
    df_train = df_train.fillna(fill_dict)
    df_test = df_test.fillna(fill_dict)
    # 3. Dealing with basement variables
    fill_dict = {
        'BsmtQual': 'Nb',
        'BsmtCond': 'Nb',
        'BsmtExposure': 'Nb',
        'BsmtFinType1': 'Nb',
        'BsmtFinType2': 'Nb'
    }
    df_train = df_train.fillna(fill_dict)
    df_test = df_test.fillna(fill_dict)
    # 4. Dropping rows for rest of the columns
    df_train = df_train.dropna()
    df_test = df_test.fillna(df_test.mode().iloc[0])
    
    return df_train, df_test
```

```python
# Step 3 - Do not apply to test
def clean_outlier(df):
    # Replace numerical outliers with column means
    cols = ['GrLivArea', 'MasVnrArea', 'TotalBsmtSF', 'GarageArea']
    for col in cols:
        col_series = df[col].copy()
        col_series[col_series>col_series.quantile(0.99)] = col_series.mean()
        df[col] = col_series
    return df
```

```python
# step 4 - Apply to test
def normalise_num_features(df_train, df_test):
    
    # Apply min-max scaling to numerical features
    df_num_train = df_train.select_dtypes(include=['int64', 'float64'])
    df_num_train = df_num_train.drop('SalePrice', axis=1)
    df_num_train = (df_num_train - df_num_train.min()) / (df_num_train.max() - df_num_train.min())
    df_num_test = df_test.select_dtypes(include=['int64', 'float64'])
    df_num_test = (df_num_test - df_num_test.min()) / (df_num_test.max() - df_num_test.min())
    df_train = df_train.drop(df_num_train.columns, axis=1)
    df_train = pd.concat([df_train, df_num_train], axis=1)
    df_test = df_test.drop(df_num_test.columns, axis=1)
    df_test = pd.concat([df_test, df_num_test], axis=1)
    
    return df_train, df_test
```

### Feature Engineering and Selection

```python
# Step 2 - Apply to test
def num_feat_select(df_train, df_test):
    
    # Select numerical features
    df_num = df_train.select_dtypes(include=['int64', 'float64'])
    # Check correlation between numerical features and target variable
    num_corr_target = df_num.corr()['SalePrice'].abs().sort_values(ascending=False)
    low_corr_cols = num_corr_target[num_corr_target<0.4].index.tolist()
    drop_cols = ['GarageCars', '1stFlrSF', 'TotRmsAbvGrd']
    df_train = df_train.drop(low_corr_cols, axis=1)
    df_train = df_train.drop(drop_cols, axis=1)
    df_test = df_test.drop(low_corr_cols, axis=1)
    df_test = df_test.drop(drop_cols, axis=1)
    
    return df_train, df_test
```

```python
# step 5 - Apply to test
def text_feat_select(df_train, df_test):
    
    # Select text features
    df_text = df_train.select_dtypes(include=['object'])
    df_text_test = df_test.select_dtypes(include=['object'])
    df_train = df_train.drop(df_text.columns, axis=1)
    df_test = df_test.drop(df_text.columns, axis=1)
    unique_count = df_text.describe().loc['unique']
    min_unique_cols = unique_count[unique_count<=5].index.tolist()
    min_unique_cols.append('Neighborhood')
    drop_cols = ['LandContour', 'LotConfig', 'LandSlope', 'ExterCond', 'BsmtCond', 'BsmtExposure', 'GarageFinish', 'Street', 'Utilities', 'LandSlope', 'CentralAir']
    for col in drop_cols:
        if col in min_unique_cols:
            min_unique_cols.remove(col)
    df_text = df_text[min_unique_cols]
    df_text_test = df_text_test[min_unique_cols]
    df_train = pd.concat([df_train, df_text], axis=1)
    df_test = pd.concat([df_test, df_text_test], axis=1)
    
    return df_train, df_test
```

```python
# Step 6 - Apply to test
def text_feat_encode(df_train, df_test):
    
    # Apply one-hot encoding to all non-numeric features
    df_text = df_train.select_dtypes(exclude=np.number)
    df_text_test = df_test.select_dtypes(exclude=np.number)
    df_train = df_train.drop(df_text.columns, axis=1)
    df_test = df_test.drop(df_text.columns, axis=1)
    df_text_dummies = pd.DataFrame()
    for col in df_text.columns:
        df_text_dummies = pd.concat([df_text_dummies, pd.get_dummies(df_text[col], drop_first=True)], axis=1)
    df_text_test_dummies = pd.DataFrame()
    for col in df_text.columns:
        df_text_test_dummies = pd.concat([df_text_test_dummies, pd.get_dummies(df_text_test[col], drop_first=True)], axis=1)
    df_train = pd.concat([df_train, df_text_dummies], axis=1)
    df_train = df_train.drop('Mix', axis=1) #Mix isn't in test set
    df_test = pd.concat([df_test, df_text_test_dummies], axis=1)
    return df_train, df_test
```

```python
# Step 7 - Don't apply to test
def target_transform(y):
    return np.log(y)
```

### Model Selection

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import KFold, cross_val_score

def model_select(df, params):
    X = df.drop('SalePrice', axis=1)
    y = target_transform(df['SalePrice'])
    kf = KFold(n_splits=10, random_state=10)
    best_model = None
    best_score = 0
    best_rmse = 0
    best_param = 0
    
    for p in params:
        model = Ridge(alpha=p)
        scores = cross_val_score(model, X, y, scoring='r2', cv=kf)
        mses = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=kf)
        rmses = [np.abs(s)**(1/2) for s in mses if s > -1]
        avg_rmses = np.mean(rmses)
        if np.mean(scores) > best_score:
            
            best_score = np.mean(scores)
            best_model = model
            best_rmse = avg_rmses
            best_param = p
        
    return best_score, best_rmse, best_model, best_param
```
